```python
import matplotlib.pyplot as plt
```

```python
#first we need to sclae the gre score with cgpa to get better insights
#for that we are doing feature scaling
from sklearn.preprocessing import MinMaxScaler



# Select the columns to be scaled
data_to_scale = df[['GRE Score', 'CGPA']]

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(data_to_scale)

# Create a new DataFrame with the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=['Scaled GRE Score', 'Scaled CGPA'], index=df.index)

# Concatenate the scaled data with the original DataFrame
df = pd.concat([df, scaled_df], axis=1)

#scatter plot using the scaled features:
plt.figure(figsize=(8, 6))
plt.scatter(df['Scaled GRE Score'], df['Scaled CGPA'], alpha=0.7, color='blue', edgecolors='black')
plt.xlabel('Scaled GRE Score', fontsize=12)
plt.ylabel('Scaled CGPA', fontsize=12)
plt.title('Relationship between Scaled GRE Score and Scaled CGPA', fontsize=14)

plt.show()
```

```python
#more will be the gre score more will be the cgpa
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['CGPA'], df['Chance of Admit '],color='pink')
plt.title('Chance of Admit vs CGPA')
plt.xlabel('CGPA', fontsize=12)
plt.ylabel('Chance of Admit', fontsize=12)
plt.show()
```

```python
#more will be the cgpa more will be the chance of admit
```

```python
plt.figure(figsize=(8, 6))
plt.pie(df['University Rating'].value_counts(), labels=df['University Rating'].unique(), autopct='%1.1f%%')
plt.title('University Rating Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(df['Research'].value_counts(), labels=df['Research'].unique(), autopct='%1.1f%%')
plt.title('Research')
plt.show()
```

```python
#56% university offer research facilities while 44% does not offer it
```

```python
# Box Plot
import seaborn as sns
plt.figure(figsize=(6,4))
sns.boxplot(x=df['University Rating'], y=df['TOEFL Score'])
plt.title('Box Plot')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(df['University Rating'], df['CGPA'], color= 'orange')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.title('Bar Plot')
plt.show()
```
