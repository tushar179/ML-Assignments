```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# read data from different formats
df = pd.read_csv('/content/heart.csv')
df.head()
```

```python

#bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='cp', hue='sex', data=df, palette='Set2')
plt.title('Bar Plot: Count of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
#Scatter plot
plt.scatter(df['age'], df['chol'])
plt.xlabel('age')
plt.ylabel('chol')
plt.title('age vs chol')
plt.show()
```

```python
# 3. Box Plot: Resting Blood Pressure by AHD
plt.figure(figsize=(8, 6))
sns.boxplot(x='age', y='restecg', data=df, palette='pastel')
plt.title('Box Plot: Resting Blood Pressure by AHD')
plt.xlabel('AHD (Heart Disease)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
# 4. Pie Chart: Distribution of Sex
sex_counts = df['sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Pie Chart: Distribution of Sex')
plt.show()
```

```python

# 5. Line Chart: Maximum Heart Rate vs. Age
plt.figure(figsize=(8, 6))
sns.lineplot(x='age', y='trestbps', hue='ca', data=df, palette='dark', marker='o')
plt.title('Line Chart: Maximum Heart Rate vs. Age')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

```python
# Histogram: Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=10, kde=True, color='skyblue')
plt.title('Histogram: Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Heatmap: Correlation Matrix
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)  # Calculate correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Matrix')
plt.show()
```

```python
# Violin Plot: Cholesterol Distribution by AHD
plt.figure(figsize=(8, 6))
sns.violinplot(x='sex', y='chol', data=df, palette='pastel')
plt.title('Violin Plot: Cholesterol Distribution by AHD')
plt.xlabel('Sex (Male & Female)')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# Pair Plot: Pairwise Relationships
sns.pairplot(df[['age', 'trestbps', 'chol', 'restecg', 'sex']], hue='sex', palette='Set2')
plt.suptitle('Pair Plot: Pairwise Relationships', y=1.02)
plt.show()
```

```python
# Bubble Plot: Age vs. Cholesterol with Oldpeak as Bubble Size
plt.figure(figsize=(10, 8))
plt.scatter(df['age'], df['chol'], s=df['oldpeak']*100, alpha=0.6, c='orange', edgecolors='black')
plt.title('Bubble Plot: Age vs. Cholesterol (Bubble Size = Oldpeak)')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```
